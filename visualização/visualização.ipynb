```python
# Listando os arquivos CSV disponíveis na pasta
csv_files = glob('./../csvs/*.csv')

# Lendo os arquivos CSV disponíveis
dfs = [pd.read_csv(f) for f in csv_files]
```

```python
# Criando função auxiliar para visualizar as funções dos dataframes
def ler_df(numero_df, df):
    print(f'\n--- Csv {numero_df} ---')
    print('\nShape:')
    print(df.shape)
    print('Head:')
    print(df.head())
    print('\nDtypes:')
    print(df.dtypes)
    print('\nInfo:')
    print(df.info())
    print('\nDescribe:')
    print(df.describe())
    print('\nVerificação de Nulos:')
    print(df.isnull().sum())

list(map(lambda args: ler_df(*args), enumerate(dfs, start=1)))
```
